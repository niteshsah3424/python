```python
#2.implement stack using list
stack = []

while True:
    print("\n1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Display")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        element = input("Enter element to push: ")
        stack.append(element)
        print("Pushed:", element)

    elif choice == 2:
        if stack == []:
            print("Stack is empty")
        else:
            print("Popped:", stack.pop())

    elif choice == 3:
        if stack == []:
            print("Stack is empty")
        else:
            print("Top element:", stack[-1])

    elif choice == 4:
        print("Stack:", stack)

    elif choice == 5:
        print("Exiting")
        break

    else:
        print("Invalid choice")
```

```python
#3.implement using class
class Stack:
    stack = []
s = Stack()

while True:
    print("\n1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Display")
    print("5. Exit")

    ch = int(input("Enter your choice: "))

    if ch == 1:
        val = input("Enter value to push: ")
        s.stack.append(val)
        print("Pushed:", val)

    elif ch == 2:
        if s.stack == []:
            print("Stack is empty")
        else:
            print("Popped:", s.stack.pop())

    elif ch == 3:
        if s.stack == []:
            print("Stack is empty")
        else:
            print("Top element:", s.stack[-1])

    elif ch == 4:
        print("Stack:", s.stack)

    elif ch == 5:
        print("Exiting program")
        break

    else:
        print("Invalid choice")
```

```python
#4.Check Balanced Parentheses
expr = input("Enter the expression: ")

stack = []

balanced = True

for char in expr:
    if char == '(':
        stack.append(char)
    elif char == ')':
        if stack == []:
            balanced = False
            break
        else:
            stack.pop()

if balanced and stack == []:
    print("Parentheses are balanced")
else:
    print("Parentheses are not balanced")
```

```python
#5.Reverse a String using Stack
string = input("Enter a string: ")

stack = []
for char in string:
    stack.append(char)

reversed_string = ""
while stack:
    reversed_string += stack.pop()
print("Reversed string:", reversed_string)
```
